```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix 
```

```python
df = pd.read_csv('uci breast cancer dataset.csv - uci breast cancer dataset.csv.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='diagnosis',data=df)
```

```python
df.value_counts('diagnosis')
```

```python
df = df.drop('id',axis=1)
```

```python
x = df.drop('diagnosis',axis=1)
```

```python
y = df['diagnosis']
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.30, random_state=101)
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc_pred = dtc.predict(x_test)
```

```python
dtc_pred 
```

```python
print(classification_report(y_test,dtc_pred))
```

```python
 print(confusion_matrix(y_test,dtc_pred))
```

```python
print(accuracy_score(y_test,dtc_pred))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(accuracy_score(y_test,rfc_pred))
```

# Grid Search CV

```python
trees = np.random.randint(15,150,30)
```

```python
max_features = ['sqrt','log2']
max_depth = np.random.randint(2,12,50)
criterion = ['gini','entropy']
min_samples_split = [2,3,4]



grids = {
        'n_estimators' : trees,
        'max_features' :max_features,
        'max_depth' : max_depth,
        'min_samples_split' : min_samples_split
       }
```

```python
from sklearn.model_selection import GridSearchCV
classifier = GridSearchCV(rfc,grids,cv=3)
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.cv_results_
```

```python
tuning_results = pd.DataFrame(classifier.cv_results_)
```
